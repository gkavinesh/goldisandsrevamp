```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the datasets
col_mat_nuw_data = pd.read_csv("col_mat_nuw_output.csv")
mon_kur_jaf_data = pd.read_csv("mon_kur_jaf_output.csv")
kan_data = pd.read_csv("kan_output.csv")
```

```python
# Separate datasets for each city
bibile_monaragala_data = mon_kur_jaf_data[mon_kur_jaf_data['Location'] == 'Bibile, Monaragala']
colombo_proper_data = col_mat_nuw_data[col_mat_nuw_data['Location'] == 'Colombo Proper']
deniyaya_matara_data = col_mat_nuw_data[col_mat_nuw_data['Location'] == 'Deniyaya, Matara']  # Corrected
jaffna_proper_data = mon_kur_jaf_data[mon_kur_jaf_data['Location'] == 'Jaffna Proper']
kandy_proper_data = kan_data[kan_data['Location'] == 'Kandy Proper']
kurunegala_proper_data = mon_kur_jaf_data[mon_kur_jaf_data['Location'] == 'Kurunegala Proper']
nuwara_eliya_proper_data = col_mat_nuw_data[col_mat_nuw_data['Location'] == 'Nuwara Eliya Proper']

```

```python

# Dictionary of datasets by city for easier iteration
datasets_by_city = {
    "Bibile, Monaragala": bibile_monaragala_data,
    "Colombo Proper": colombo_proper_data,
    "Deniyaya, Matara": deniyaya_matara_data,
    "Jaffna Proper": jaffna_proper_data,
    "Kandy Proper": kandy_proper_data,
    "Kurunegala Proper": kurunegala_proper_data,
    "Nuwara Eliya Proper": nuwara_eliya_proper_data,
}

# Initialize a dictionary to store the cleaned dataframes for each city
cleaned_data_per_city = {}

for city, city_data in datasets_by_city.items():
    # Make a copy of the dataframe to avoid SettingWithCopyWarning
    city_data_clean = city_data.copy()

    # Identify rows with NaN values in 'HCHO Reading'
    nan_rows_indices = city_data_clean[city_data_clean['HCHO Reading'].isna()].index

    # If there are NaN values, display a few of them
    if not nan_rows_indices.empty:
        print(f"City: {city} - Rows with NaN values in 'HCHO Reading' before imputation:")
        print(city_data_clean.loc[nan_rows_indices].head())

    # Calculate the median of the 'HCHO Reading' column, excluding NaN values
    median_value = city_data_clean['HCHO Reading'].median()

    # Impute NaN values in the 'HCHO Reading' column with its median
    city_data_clean.loc[nan_rows_indices, 'HCHO Reading'] = median_value

    # Store the imputed dataframe in the dictionary
    cleaned_data_per_city[city] = city_data_clean

    # After imputation, print the rows that had NaN, now replaced with the median
    if not nan_rows_indices.empty:
        print(f"\nCity: {city} - The same rows after imputation with median ({median_value}):")
        print(city_data_clean.loc[nan_rows_indices].head())
        print("-"*50 + "\n")

```

```python
# Print the number of columns and the number of missing values after imputation for each city
for city, city_data in cleaned_data_per_city.items():
    print(f"{city} null values:")
    print(city_data.isnull().sum())
    print()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the function for Winsorization
def winsorize(data, lower_pct=0.05, upper_pct=0.95):
    lower_bound = data.quantile(lower_pct)
    upper_bound = data.quantile(upper_pct)
    winsorized_data = data.clip(lower=lower_bound, upper=upper_bound)
    return winsorized_data

# Assuming you have individual datasets for each city

# Create a dictionary to store winsorized data for each city
winsorized_data_per_city = {}

# Apply Winsorization to the 'HCHO Reading' column for each city's dataset
for city, data in [("Bibile, Monaragala", bibile_monaragala_data),
                   ("Colombo Proper", colombo_proper_data),
                   ("Deniyaya, Matara", deniyaya_matara_data),
                   ("Jaffna Proper", jaffna_proper_data),
                   ("Kandy Proper", kandy_proper_data),
                   ("Kurunegala Proper", kurunegala_proper_data),
                   ("Nuwara Eliya Proper", nuwara_eliya_proper_data)]:
    winsorized_data_per_city[city] = data.copy()  # Copy the data to avoid modifying the original DataFrame
    winsorized_data_per_city[city]['HCHO Reading Winsorized'] = winsorize(data['HCHO Reading'])

# Plot boxplots before and after Winsorization for each city
for city, data in winsorized_data_per_city.items():
    plt.figure(figsize=(10, 6))
    # Before Winsorization
    plt.subplot(1, 2, 1)
    before_boxplot = data.boxplot(column='HCHO Reading', grid=False)
    plt.title(f'Boxplot of HCHO Reading Before Winsorization - {city}')
    plt.xlabel('Location')
    plt.ylabel('HCHO Reading')
    plt.xticks(rotation=45)

    # After Winsorization
    plt.subplot(1, 2, 2)
    after_boxplot = data.boxplot(column='HCHO Reading Winsorized', grid=False)
    plt.title(f'Boxplot of HCHO Reading After Winsorization - {city}')
    plt.xlabel('Location')
    plt.ylabel('HCHO Reading (Winsorized)')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
```

```python
for city_name, dataset in datasets_by_city.items():
    # Display city name
    print(f"{city_name}:")

    # Save original dates for comparison
    original_current_date = dataset['Current Date'].head().copy()
    original_next_date = dataset['Next Date'].head().copy()

    # Convert date columns to datetime format, using loc to avoid SettingWithCopyWarning
    dataset.loc[:, 'Current Date'] = pd.to_datetime(dataset['Current Date'], format='%d-%m-%Y')
    dataset.loc[:, 'Next Date'] = pd.to_datetime(dataset['Next Date'], format='%d-%m-%Y')

    # Concatenate original and formatted dates for side by side comparison
    comparison_table = pd.concat([
        original_current_date.rename('Original Current Date'),
        dataset['Current Date'].head().rename('Formatted Current Date'),
        original_next_date.rename('Original Next Date'),
        dataset['Next Date'].head().rename('Formatted Next Date')
    ], axis=1)

    # Display the table
    print(comparison_table)
    print("-----------------------------------------------------\n")
```
